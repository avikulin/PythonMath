```python
from matplotlib import pyplot as plt
from matplotlib import ticker as plt_ticker
import numpy as np
```

```python
#preparing the sample data to plot
x_space = np.linspace(-5,5,1000)
z_func = lambda z, x: z - x
z0_line = map (lambda x: z_func(0,x), x_space)
z1_line = map (lambda x: z_func(1,x), x_space)
z2_line = map (lambda x: z_func(2,x), x_space)
z5_line = map (lambda x: z_func(5,x), x_space)
```

```python
%matplotlib inline

#preparing the canvas with title and coordinate system with limits(x=[-5,5]m y=[-5,7])
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(1, 1, 1)
plt.grid(which="both")
plt.ylim(bottom=-5, top=7)
plt.xlim(left=-5, right=6)

plt.title("Sample mutual distribution graph", fontdict= {"color":  "black","weight": "bold", "size": 16})

#drawing X and Y axes with arrows
plt.hlines(y=0, xmin=-5, xmax=5, linewidth = 2, color="k")
plt.text(x=5, y=0.25, s="$X$", fontdict= {"color":  "black","weight": "bold", "size": 16})

plt.vlines(x=0, ymin=-5, ymax=6, linewidth = 2, color="k")
plt.text(x=0.25, y=6, s="$Y$", fontdict= {"color":  "black","weight": "bold", "size": 16})

plt.arrow(x=4.92, y= 0, dx= 0.1, dy=0, width=0.002, color="k", clip_on=False, head_width=0.12, head_length=0.25)
plt.arrow(x=0.01, y=5.92, dx=0, dy=0.1, width=0.002, color="k", clip_on=False, head_width=0.12, head_length=0.25)
    
# setting major (width=1) and minor(width=0.5) ticks on both axes
major_ticks = plt_ticker.MultipleLocator(base=1.0)
minor_ticks = plt_ticker.MultipleLocator(base=0.5)
ax.xaxis.set_minor_locator(minor_ticks)
ax.xaxis.set_major_locator(major_ticks)
ax.yaxis.set_minor_locator(minor_ticks)
ax.yaxis.set_major_locator(major_ticks)

# Fill the integration areas
plt.fill_between(x=x_space[500:600], y1=z5_line[500:600],y2= np.zeros(100), facecolor="lightsalmon", edgecolor="orangered", linewidth=1)
plt.text(x=0.375, y=3.0, s="$A$",fontdict= {"color":  "black","weight": "bold", "size": 16})

plt.fill_between(x=x_space[600:700], y1=z5_line[600:700],y2= np.zeros(100), facecolor="lightgreen", edgecolor="forestgreen", linewidth=1)
plt.text(x=1.375, y=2.0, s="$B$",fontdict= {"color":  "black","weight": "bold", "size": 16})

# drawing the Z functions as integration limits
plt.plot(x_space, z0_line, "r-", linewidth=0.5)
plt.annotate('z=0', xy=(-1, 1), xytext=(-2.5, -0.5), arrowprops=dict(facecolor="red", arrowstyle="->"), )

plt.plot(x_space, z1_line, "r-", linewidth=0.5)
plt.annotate('z=1', xy=(1.5, -0.5), xytext=(0.4, -1.8), arrowprops=dict(facecolor="red", arrowstyle="->"), )

plt.plot(x_space, z2_line, "r-", linewidth=0.5)
plt.annotate('z=2', xy=(-3.5, 5.5), xytext=(-2.5, 6.5), arrowprops=dict(facecolor="red", arrowstyle="->"), )

plt.plot(x_space, z5_line, "r--", linewidth=1.5)
plt.annotate('z=2.5', xy=(-1, 6), xytext=(-2, 5), arrowprops=dict(facecolor="red", arrowstyle="->"), )

#drafing X integration limit
plt.vlines(x=2, ymin=-5, ymax=6,  linewidth = 2, color="brown")
plt.text(x=2.1, y=3.2, s="$X=2$")

plt.vlines(x=1, ymin=-5, ymax=6,  linewidth = 1.5, color="gray", linestyles="dashed")
plt.text(x=1.1, y=4.2, s="$X=1$")

#finalizing the canvas to show
plt.show()
```
